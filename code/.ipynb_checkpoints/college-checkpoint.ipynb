## Assignment

- Create a for loop that plots the histogram of sample size, median, employed, full time, unemployment rate, men and women (remember to use the exact column titles)

```python
col=['Sample_size', 'Median', 'Employed', 'Full_time_year_round', 'Unemployment_rate', 'Men', 'Women']

figs={}
for idx,x in enumerate(col):
    figs[idx]=plt.figure()
    figs[idx]=figs[idx].add_subplot(111)
    df[x].plot(kind='hist',title=x)
```

## Assignment

- Make a scatter matrix of sample size and median
- Make a scatter matrix of sample size, median and unemployment rate


```python
from pandas.tools.plotting import scatter_matrix
df_ssm=df[['Sample_size','Median']]
scatter_matrix(df_ssm, alpha=0.2, figsize=(6, 6))
```

```python
df_ssmu=df[['Sample_size','Median','Unemployment_rate']]
scatter_matrix(df_ssmu, alpha=0.2, figsize=(6, 6))
```

## Assignment

- Make a bar plot of majors vs "ShareWomen" (you should run different slices since there are many different majors)


```python
attr = df['Major_category'].unique()
for e in attr:
    df2=df[df['Major_category']==e]
    df2=df2[['Major','ShareWomen']]
    df2.plot(kind='bar',title=e)
```

```python
d=df.groupby('Major_category')['ShareWomen'].mean()
d.plot(kind='bar')
```
