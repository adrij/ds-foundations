```python
#Look at the age distribution.
import matplotlib
import seaborn as sns
%matplotlib inline
data_th['Age'].describe()
sns.countplot(x='Age' , data=data_th,order=['18 - 29', '30 - 44', '45 - 59', '60+'])
```

```python
#Write a function that approximates each respondent's age given the ranges and apply it to all cells (remember type conversion).
import re
def func_age(bin):
    if pd.isnull(bin):
        return None
    lst=re.findall(r'\d+', bin)
    s=0
    for i in lst:
        s=s+int(i)
    return s/len(lst)
data_th['Age_num']=data_th.apply(lambda x: func_age(x['Age']), axis=1)
print (data_th.tail(20))
```

```python
#Examine the summary statistics of the age column now.
data_th['Age_num'].describe()
```

```python
#Look at the income distribution.
print(data_th['How much total combined money did all members of your HOUSEHOLD earn last year?'].describe())
print (data_th['How much total combined money did all members of your HOUSEHOLD earn last year?'].unique())
```

```python
#Write a function that approximates each respondent's income given the ranges and apply it to all cells (remember type conversion).

def func_income(bin):
    if pd.isnull(bin):
        return None
    lst=re.findall(r'\d+', bin)
    s=0
    if len(lst)==0:
        return None
    for i in range(0,len(lst),2):
        if i==0:
            s=s+int(lst[i])
        else:
            s=s+int(lst[i])+1
    return s/len(lst)*2*1000
data_th['Income_num']=data_th.apply(lambda x: func_income(x['How much total combined money did all members of your HOUSEHOLD earn last year?']), axis=1)
print (data_th.tail(20))
```

```python
#Examine the summary statistics of the income column now.
data_th['Income_num'].describe()
```

```python
#Look at the distance traveled for lower income (< 50,000) vs. higher income (>150,000) respondents.
print('Lower income:')
lower_inc=data_th[data_th['Income_num']<50000]
print(lower_inc['How far will you travel for Thanksgiving?'].value_counts()/len(lower_inc))
higher_inc=data_th[data_th['Income_num']>150000]
print('Higher income:')
print(lower_inc['How far will you travel for Thanksgiving?'].value_counts()/len(higher_inc))
```

```python
#Use the pivot_table function to examine what ages and incomes are more likely to "attend a Friendsgiving" or "meetup with hometown friends.
def func_yn(yn):
    if yn=='Yes':
        return True
    if yn=='No':
        return False
    else:
        return None 
data_th['Friendsgiving']=data_th.apply(lambda x: func_yn(x['Have you ever attended a "Friendsgiving?"']),axis=1)
data_th['meetup']=data_th.apply(lambda x: func_yn(x['Have you ever tried to meet up with hometown friends on Thanksgiving night?']),axis=1)
data_th['Friend_meetup']=data_th['Friendsgiving']|data_th['Friend_meetup']
#Ratio of True values
pd.pivot_table(data_th, values='Friend_meetup',index=['Age_num'], columns='Income_num', aggfunc=lambda x: len(x[x==True])/len(x))
```
